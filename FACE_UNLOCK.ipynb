```python
from deepface.basemodels import Facenet
import glob
import warnings
warnings.filterwarnings('ignore')
from deepface.commons import functions
import faiss
import numpy as np
```

```python
model = Facenet.loadModel()
```

```python
representations = []
for i in glob.glob('/home/ubuntu/Downloads/my images/*'):
    img = functions.preprocess_face(img=i,target_size=(160,160),enforce_detection=False)
    embedding = model.predict(img)[0,:]
    
    representation = []
    representation.append(i)
    representation.append(embedding)
    representations.append(representation)
    
    
    
```

```python
representations[0][1]
```

```python
embedding.shape
```

```python
embeddings = []
for i in range(0,len(representations)):
    embedding = representations[i][1]
    embeddings.append(embedding)
embeddings = np.array(embeddings,dtype='f')    
```

```python
embeddings.shape
```

# initializing faiss index

```python
dimensions = 128

index = faiss.IndexFlatL2(dimensions)
index.add(embeddings)
```

```python
#saving our model
faiss.write_index(index,'vector1.index')
```
